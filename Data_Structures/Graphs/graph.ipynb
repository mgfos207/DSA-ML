# Graphs

The purpose of this notebook is to explore the graph data structures and understand the various forms it takes, as well as how to utilize them.

```python
from graph_data_structure import DirectedGraph
```

### Directed Non-Weighted Graph

```python
num_vertices = 6
edges = [(0, 1), (1, 2), (2, 0), (2, 1), (3, 2),
        (4, 5), (5, 4)]
graph_weighted = False
direct_graph_obj = DirectedGraph(num_vertices, graph_weighted)
direct_graph_obj.add_to_adj_list(edges)
```

```python
direct_graph_obj.adj_list
```

```python
direct_graph_obj.visualize_graph()
```

### Directed Weighted Graph

```python
num_vertices = 6
edges = [(0, 1, 6), (1, 2, 7), (2, 0, 5), (2, 1, 4), (3, 2, 10),
        (4, 5, 1), (5, 4, 3)]
graph_weighted = True
weighted_direct_graph_obj = DirectedGraph(num_vertices, graph_weighted)
weighted_direct_graph_obj.add_to_adj_list(edges)
```

```python
weighted_direct_graph_obj.adj_list
```

```python
weighted_direct_graph_obj.visualize_graph()
```
